```python
import cv2
import numpy
import matplotlib.pyplot as plt

img=cv2.imread("img.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

blur_img1=cv2.blur(img,(10,10))

blur_img2=cv2.GaussianBlur(img,(5,5),8)

blur_img3=cv2.medianBlur(img,3)

plt.subplot(2,2,1),plt.imshow(img),plt.title("Original Image")
plt.subplot(2,2,2),plt.imshow(blur_img1),plt.title("Box Blur")
plt.subplot(2,2,3),plt.imshow(blur_img2),plt.title("Gaussian Blur")
plt.subplot(2,2,4),plt.imshow(blur_img3),plt.title("Median Blur")
```
